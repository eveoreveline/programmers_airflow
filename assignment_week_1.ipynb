## Dev 생성

```python
# 라이브러리 로드
import pandas as pd
```

```python
# 데이터 프레임 생성
df = pd.read_csv("kaggle_netflix.csv")
```

```python
df.head()
```

```python
# 사용할 컬럼으로 dev 생성
dev = df[["id", "title", "type", "release_year", "age_certification", "runtime", "genres", "production_countries"]]
dev.head()
```

### 데이터 전처리

```python
dev['genre_1'] = dev.genres.str.split(',').str[0]
dev['genre_2'] = dev.genres.str.split(',').str[1]
dev['genre_3'] = dev.genres.str.split(',').str[2]
dev.head()
```

```python
dev['production_countries_1'] = dev.production_countries.str.split(',').str[0]
dev['production_countries_2'] = dev.production_countries.str.split(',').str[1]
dev.head()
```

```python
import re
dev["genre_1"] = dev["genre_1"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dev["genre_2"] = dev["genre_2"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dev["production_countries_1"] = dev["production_countries_2"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
dev.head()
```

## Analytics 생성

```python
number_of_contents=dev.count()
number_of_contents
```

```python
movies_per_genres=dev['genre_1'].value_counts()
movies_per_genres
```

```python
shows_per_release_year = dev.groupby(['release_year', 'type']).count()
shows_per_release_year
```

```python
shows_per_production_countries = dev.groupby(['production_countries_1', 'type']).count()
shows_per_production_countries
```

```python
age_certification_distribution=dev['age_certification'].value_counts()
age_certification_distribution
```

```python
import seaborn as sns
sns.countplot(x=dev['age_certification'], order=age_certification_distribution.index)
```

```python
runtime_distribution=dev['runtime'].value_counts()
runtime_distribution
```

```python
import matplotlib.pyplot as plt
sns.distplot(dev['runtime'])
plt.show()
```
